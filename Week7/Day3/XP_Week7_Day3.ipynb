ðŸŒŸ Exercise 2 : Descriptive Statistics
Instructions
Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.
Sample dataset in code:


data = [12, 15, 13, 12, 18, 20, 22, 21]

```python
import scipy.stats as stats
import scipy as sp
import numpy as np


data = [12, 15, 13, 12, 18, 20, 22, 21]

mean_scipy = stats.tmean(data)
median_scipy = np.median(data)  # SciPy does not have a separate median function
variance_scipy = stats.tvar(data)
std_dev_scipy = stats.tstd(data)

print(mean_scipy)
print(median_scipy)
print(variance_scipy)
print(std_dev_scipy)
```


ðŸŒŸ Exercise 3 : Understanding Distributions
Instructions
Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random samples from the normal distribution
data1= np.random.normal(50, 10, 1000)


mean_scipy1 = stats.tmean(data1)
median_scipy1 = np.median(data1)  # SciPy does not have a separate median function
variance_scipy1 = stats.tvar(data1)
std_dev_scipy1 = stats.tstd(data1)

print(mean_scipy1)
print(median_scipy1)
print(variance_scipy1)
print(std_dev_scipy1)


sns.histplot(data1, kde=True)  # kde=True adds a kernel density estimate curve
plt.title('Normal Distribution (Mean=50, Std Dev=10)')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()


```

ðŸŒŸ Exercise 4 : T-Test Application
Instructions
Perform a T-test on two sets of randomly generated data. Hint: Use numpy.random.normal() to generate the datasets.

```python
import numpy as np
import scipy.stats as stats
set1 = np.random.normal(loc=50, scale=10, size=10)
set2 = np.random.normal(loc=50, scale=12, size=10)

print(set1)
print(set2)

t_test= stats.ttest_ind(set1,set2)
print(t_test)
```
