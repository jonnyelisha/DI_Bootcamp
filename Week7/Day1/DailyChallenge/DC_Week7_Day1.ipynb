Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities. The temperatures should range from -5 to 35 degrees.

Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns.

```python

!pip install faker
!pip install numpy
!pip install pandas


import numpy as np
import pandas as pd
from faker import Faker
fake = Faker()
num_cities = 10
num_months = 12
temperature_data = np.random.uniform(low=-5, high=35, size=(num_cities, num_months))

cities = [fake.city() for _ in range(num_cities)]
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']

temperature_df = pd.DataFrame(data=temperature_data, index=cities, columns=months)
temperature_df
```

```python
annual_avg_temp = temperature_df.mean(axis=1)
annual_avg_temp
```

```python
city_highest_avg_temp = annual_avg_temp.idxmax()
city_lowest_avg_temp = annual_avg_temp.idxmin()
print(city_highest_avg_temp)
print(city_lowest_avg_temp)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
for city in cities:
    plt.plot(months, temperature_df.loc[city], marker='o', label=city)

plt.xlabel('Month')
plt.ylabel('Average Temperature (Â°C)')
plt.title('Monthly Average Temperature for Each City')
plt.xticks(rotation=45, ha='right')

plt.grid(which='both', axis='both', linestyle='--', linewidth=0.5)
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```
