```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df  = pd.read_csv('world_happiness_report_2019.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
plt.scatter(df['Score'], df['Social support'])
plt.title("Social support across Score Scatter Plot")
plt.xlabel("Score")
plt.ylabel("Social support")

slope, intercept, r_value, p_value, std_err = stats.linregress(df['Score'], df['Social support'])

line = slope * df['Score'] + intercept

# Plot the line of best fit
plt.plot(df['Score'], line, color='red', linewidth=2, label='Line of Best Fit')

# Add a legend
plt.legend()

# Show plot
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(20, 20))
# Plot GDP per Capita
axs[0].bar(df['Country or region'], df['GDP per capita'], color='blue')
axs[0].set_title('GDP per Capita across Countries')
axs[0].set_xlabel('Country')
axs[0].set_ylabel('GDP per Capita')
axs[0].tick_params(axis='x', rotation=90)

# Plot Healthy Life Expectancy
axs[1].bar(df['Country or region'], df['Healthy life expectancy'], color='green')
axs[1].set_title('Healthy Life Expectancy across Countries')
axs[1].set_xlabel('Country')
axs[1].set_ylabel('Healthy Life Expectancy')
axs[1].tick_params(axis='x', rotation=90)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(20, 15))

# Bar plot for GDP per Capita
ax1.bar(df['Country or region'], df['GDP per capita'], color='blue', alpha=0.6, label='GDP per Capita')
ax1.set_xlabel('Country or region')
ax1.set_ylabel('GDP per capita', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.tick_params(axis='x', rotation=90)

# Create a secondary y-axis for Healthy Life Expectancy
ax2 = ax1.twinx()
ax2.plot(df['Country or region'], df['Healthy life expectancy'], color='green', marker='o', label='Healthy Life Expectancy')
ax2.set_ylabel('Healthy life expectancy', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Add titles and legends
plt.title('GDP per Capita and Healthy Life Expectancy across Countries')
fig.tight_layout()
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```
