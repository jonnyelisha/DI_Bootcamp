```python
import pandas as pd
import numpy as np
datascience_df = pd.read_csv('/content/datascience_salaries.csv')
datascience_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#Normalize the salary column using Min_Max_Scale
scalar = MinMaxScaler()
datascience_df['normal_salary'] = scalar.fit_transform(datascience_df[['salary']])
datascience_df.head()
```
