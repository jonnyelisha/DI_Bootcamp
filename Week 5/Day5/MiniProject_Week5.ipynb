```python
#DailyChallenge
import pandas as pd
mental_health_df = pd.read_csv('mental_health_survey.csv')
mental_health_df.head()

#This data is filled with qualitiative data, so given that it is qualitative it won't be possible to have a distribution but it would be able to make it binary for family_history and treatment
```

```python
mental_health_df = mental_health_df.drop(columns='Timestamp')
mental_health_df =mental_health_df.drop(columns='comments')
mental_health_df.head()
```

```python
mh_columns = mental_health_df.columns
mh_columns
```

```python
# Creating another column for age groups
age_bins = [0, 20, 30, 40, 50, 60, 100]
age_labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61+']

# Create a new column 'Age Group' using pd.cut()
mental_health_df['Age Group'] = pd.cut(mental_health_df['Age'], bins=age_bins, labels=age_labels, right=False)
mental_health_df
```

```python
import matplotlib.pyplot as plt
mental_health_df.groupby(['Age Group', 'mental_health_consequence'])['mental_health_consequence'].count().unstack().plot(kind='bar')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Distribution of Mental Health Conditions Among Age Groups')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

According to the graph, it looks skewed to the right but it also looks fairly normal.

```python
# Filter DataFrame for mental health consequences and tech company
filtered_df = mental_health_df[(mental_health_df['mental_health_consequence'] == 'Yes') & (mental_health_df['tech_company'] == 'Yes')]
ordered_df = filtered_df.sort_values('Age Group')
ordered_df
```

```python
# Total Group Count: Count the number of people in each age group
age_group_counts = ordered_df['Age Group'].value_counts()

print(age_group_counts)
```

```python
counts = ordered_df.groupby(['Gender', 'mental_health_consequence'])['mental_health_consequence'].count().unstack()

# Create a bar plot
counts.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of Mental Health Consequences by Gender')
plt.xticks(rotation=0)
plt.legend(title='Mental Health Consequence')
plt.tight_layout()
plt.show()
```

```python
#Identify the countries will the highest reported rates of mental health issues of the tech industry

mh_tech_df = mental_health_df[(mental_health_df['mental_health_consequence'] == 'Yes') & (mental_health_df['tech_company'] == 'Yes')]

# Group by country and count 'Yes' responses
country_counts = mh_tech_df['Country'].value_counts()

# Get the top 5 countries
top_5_countries = country_counts.head(5)

# Print the top 5 countries and their counts
print(top_5_countries)
```

```python
#Countries with the lowest reported rates of mental_health_concerns.
bottom_5_countries = country_counts.tail(5)
print(bottom_5_countries)
```

```python
#Explanation: It is interesting to see that the North American countries ( Canada, US) have higher reported rates of mental health concerns within the Tech industry than countries in Europe.
```
